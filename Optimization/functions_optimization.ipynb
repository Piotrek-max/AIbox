# Functions Python


```python
import numpy as np

a = np.random.randint(0,10,size=(100000,100000))
c = np.array(a, order='F')
f = np.array(a, order='C')

import time
start = time.time()
np.sum(c, axis=0)
end = time.time()
print(end - start)
start = time.time()
np.sum(f, axis=0)
end = time.time()
print(end - start)

```
